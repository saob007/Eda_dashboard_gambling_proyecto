## 1.3. Generar las consultas a realizar sobre la base de datos

El primer objetivo es identificar los jugadores que hicieron su primer depósito FTD y filtrar aquellos
que cumplen con la condición de CPA (total de depósitos mayor a 100 USD. Para ello se construyen las siguientes consulta SQL.

La primera consulta se asigna a una variable llamada “query1” en Python para poder ser utilizada luego como argumento en la creación del dataframe. El código utilizado en esta primera consulta es el siguiente.

```python
# Se genera la query para filtrar a los jugadores que hicieron su primer depósito FTD y 
# que cumplen con la condición de CPA (total de depósitos mayor a 100 USD
query1 = """
    -- Este primer CTE identifica el id para cada jugador y la fecha más lejos en la tabla de depósitos
    -- Adicionalmente, agrupa por el ID del jugador para que solo exista un único registro por jugador.
    WITH PlayerFirstDeposit AS (
        SELECT
            d.player_id,
            MIN(d.deposit_date) AS first_deposit_date
        FROM
            deposits d
        GROUP BY
            d.player_id
    ),
    
    -- Este segundo CTE cálcula el total de los depositos por jugador y los agrupa por el ID de cada uno.
    PlayerDepositSummary AS (
        SELECT
            d.player_id,
            SUM(d.deposit_amount) AS total_deposit_amount
        FROM
            deposits d
        GROUP BY
            d.player_id
    )
    
    -- Esta es la consulta principal donde se seleccionan 4 columnas de la tabla "player"
    -- 1 columna del CTE "PlayerFirstDeposit", 1 columna del CTE "PlayerDepositSummary"
    -- y 2 columnas de la tabla "traffic_sources", luego se usa un JOIN para relacionar 
    -- las fechas del primer depósito con los totales de depósitos de cada jugador. Otro JOIN
    -- para agregar la información básica del jugador y un tercer JOIN para unir todo esto con
    -- los datos de adquisición.
    SELECT
        p.player_id,
        p.name,
        p.country,
        p.registration_date,
        fd.first_deposit_date,
        ps.total_deposit_amount,
        ts.cost_of_acquisition,
        ts.trafficker
    FROM
        PlayerFirstDeposit fd
    JOIN PlayerDepositSummary ps ON fd.player_id = ps.player_id
    JOIN players p ON fd.player_id = p.player_id
    JOIN traffic_sources ts ON p.player_id = ts.player_id
    -- La condición es que se filtren los jugadores cuyo total de los depósitos es mayor a 100
    WHERE
        ps.total_deposit_amount > 100;
"""
```

El segundo objetivo es generar un reporte de la cantidad de FTD por mes y de CPA por mes.

La segunda consulta se asigna a una variable llamada “query2” en Python para poder ser utilizada luego como argumento en la creación del dataframe. El código utilizado en esta primera consulta es el siguiente.

```python
# Se genera la query para generar una tabla con la cantidad de FTD por mes y de CPA por mes,
# además de la representatividad del CPA según el FTD, el costo total de adquisición mensual
# y el costo promedio de adquisición equivalente al costo total mensual sobre el CPA por mes. 
query2 = """
    -- Este CTE encuentra la primera fecha del primer depósito para cada jugador en la tabla "deposits"
    WITH PlayerFirstDeposit AS (
        SELECT
            d.player_id,
            MIN(d.deposit_date) AS first_deposit_date
        FROM
            deposits d
        GROUP BY
            d.player_id
    ),
    -- Este CTE calcula el total de los depositos por jugador
    PlayerDepositSummary AS (
        SELECT
            d.player_id,
            SUM(d.deposit_amount) AS total_deposit_amount
        FROM
            deposits d
        GROUP BY
            d.player_id
    ),
    -- Este CTE agrupa los datos por mes y calcula el número de FTD y de CPA (depósitos mayores a 100)
    MonthlyFTDCPA AS (
        SELECT
            strftime('%Y-%m', fd.first_deposit_date) AS month,
            COUNT(DISTINCT fd.player_id) AS ftd_count,
            SUM(CASE WHEN ps.total_deposit_amount > 100 THEN 1 ELSE 0 END) AS cpa_count
        FROM
            PlayerFirstDeposit fd
        LEFT JOIN PlayerDepositSummary ps ON fd.player_id = ps.player_id
        GROUP BY
            strftime('%Y-%m', fd.first_deposit_date)
    ),
    -- Este CTE suma el total de los costos de adquisión para cada mes desde la tabla "traffic_sources"
    MonthlyCPAData AS (
        SELECT
            strftime('%Y-%m', ts.registration_date) AS month,
            SUM(ts.cost_of_acquisition) AS total_acquisition_cost
        FROM
            traffic_sources ts
        GROUP BY
            strftime('%Y-%m', ts.registration_date)
    )
    -- Esta es la consulta principal donde se seleccionan 3 columnas de la CTE MonthlyFTDCPA, 1 columna de la CTE
    -- MonthlyCPAData y además, se calculan 2 nuevas métricas que son: 1) El porcentaje de CPA que equivale a la 
    -- cantidad de CPA entre la cantidad de FTD por 100, 2) Costo promedio CPA que equivale al costo total de 
    -- adquisición entre la cantidad de CPA. 
    SELECT
        mftd.month,
        mftd.ftd_count,
        mftd.cpa_count,
        ROUND(CAST(mftd.cpa_count AS FLOAT) / NULLIF(mftd.ftd_count, 0) * 100, 2) AS cpa_percentage,
        mcpa.total_acquisition_cost,
        CASE
            WHEN mftd.cpa_count > 0 THEN mcpa.total_acquisition_cost / mftd.cpa_count
            ELSE 0
        END AS avg_cpa
    FROM
        MonthlyFTDCPA mftd
    LEFT JOIN MonthlyCPAData mcpa ON mftd.month = mcpa.month
    ORDER BY
        mftd.month;
"""
```

Mientras tanto, la Fig2 muestra que en el dataframe "df_ftd_cva_report", la única variable con datos atípicos, un poco más pronunciados, es la variable `avg_cpa`.
- Para identificar la cantidad exacta y los registros con valores atípicos se procede a determinar los límites máximos y mínimos de movilidad para estas dos variables. Siguiendo el método IQR 1.5.

### Identificación de registros con valores atípicos para "df_ftd_cva100_filter"

```python
# Se determinan los cuartiles 1 y 3 para la variable "total_deposit_amount".
q1 = df_ftd_cva100_filter['total_deposit_amount'].quantile(0.25)
q3 = df_ftd_cva100_filter['total_deposit_amount'].quantile(0.75)

# Se calcula el rango intercuartil.
iqr = q3 - q1

# Se identifican los límites máximo y mínimo.
limite_inferior = max(0, q1 - iqr * 1.5)  # Como la variable "total_deposit_amount" no puede tomar valores negativos,
                                          # se establece su valor en 0 si el límite calculado es negativo.
limite_superior = q3 + iqr * 1.5

# Se identifican los registros que se encuentren fuera de este rango.
mask = (df_ftd_cva100_filter['total_deposit_amount'] < limite_inferior) | \
       (df_ftd_cva100_filter['total_deposit_amount'] > limite_superior)  # Máscara de filtrado.

# Se crea un subconjunto de datos con solo los registros que tienen valores atípicos.
df_ftd_cva100_filter_outliers = df_ftd_cva100_filter[mask]

# Se imprimen los resultados de los límites y la cantidad de registros con valores atípicos.
print(f'El límite inferior para la variable "total_deposit_amount" es: {limite_inferior}')
print(f'El límite superior para la variable "total_deposit_amount" es: {limite_superior}')
print(f'Existen {mask.sum()} registros con valores atípicos')
```

### Identificación de registros con valores atípicos para "df_ftd_cva_report"

## 3.3 Se calculan las discrepancias entre los resultados y las expectativas

## 4.2 Histórico del dinero depositado

Se utilizará un gráfico de linea de tiempo para entender mejor como ha sido el comportamiento de los depósitos a través de los diferentes periodos registrados. Para lograr esto, es necesario hacer una consulta a la base de datos que permita obtener los datos de año y mes de la tabla “deposits” y calcular el total de los depósitos realizados cada periodo mensual y ordenando la tabla resultado de manera descendente según el mes.

Existe la necesidad de volver a realizar la conexión a la base de datos, como se ha hecho previamente cuando se decide importar datos para el análisis. definiendo la consulta SQL y guardandola en memoria como una variable llamada “query4”.

```python
# Se realiza la onexión a la base de datos
conn = sqlite3.connect('/kaggle/input/internal-deposits-costs-database/internal_deposits_costs_database.db') 

# Se define la consulta SQL para calcular el monto total de depósitos por mes, formateando la fecha como 'YYYY-MM',
# y se almacena en una variable llamada "query4".
query4 = """
SELECT
    strftime('%Y', deposit_date) || '-' || strftime('%m', deposit_date) AS month,  -- Se obtiene el mes y el año como un único valor formateado.
    SUM(deposit_amount) AS total_deposit_amount  -- Se calcula la suma total de los montos de depósito por mes.
FROM
    deposits  -- Se toma la información de la tabla "deposits".
GROUP BY
    month  -- Se agrupan los resultados por el mes calculado.
ORDER BY
    month DESC;  -- Se ordenan los resultados en orden descendente según el mes.
"""

# Se ejecuta la consulta y se cargan los resultados en un DataFrame de pandas
df_deposit_summary = pd.read_sql(query4, conn)

# Se cierra la conexión
conn.close()
```

Se ordena el dataframe por la columna “month” en orden ascendente y se hace asignan nuevos indexes al dataframe, eliminando los anteriores que venían con la consulta. Anteriormente, en la consulta se había indicado que los resultados se ordenasen de manera descendente pero la intención real era ordenarlos de manera ascendente. ## 4.3 Jugador con mayor depósito realizado

Es fundamental identificar a los mejores clientes (jugadores) para desarrollar y evaluar estrategias de fidelización basadas en casos exitosos. Por esta razón, se analizará el top 10 de jugadores que han realizado los mayores depósitos y se evaluará qué tan significativa es su diferencia respecto a los demás jugadores. Para esto, se creará un gráfico de barras horizontal con Seaborn que permita graficar el ID del jugador vs el total depositado a lo largo del histórico de datos.


Existe la necesidad de volver a realizar la conexión a la base de datos, como se ha hecho previamente cuando se decide importar datos para el análisis. Esto fundamentalmente a que se requieren importar todos los datos la tabla depósitos sin realizar agregaciones como se realizó en la consulta de la sección anterior.

```python
# Se realiza la conexión a la base de datos SQLite
conn = sqlite3.connect('/kaggle/input/internal-deposits-costs-database/internal_deposits_costs_database.db') 

# Se define la consulta SQL
query5 = """
    SELECT *
    FROM
        deposits;
"""
# Se ejecuta la consulta y se carga los resultados en un DataFrame de pandas
df_deposits = pd.read_sql(query5, conn)

# Se cierra la conexión
conn.close()
```

se agrupan los registros por el ID de los jugadores y se calcula el valor del total depositado para cada uno, asignando esta transformación a un nuevo dataframe llamado “player_deposits”. Para ello, se utiliza la función “groupby” y se indican la agregación de suma a realizar sobre la columna “deposit_amount”. Adicionalmente, se ordenan los registros por la columna “deposit_amount” de manera descendente para que los jugadores con más dinero depositado queden en la parte superior de los registros.

```python
# Se agrupan los datos del DataFrame "df_deposits" por la columna "player_id" para calcular el valor total depositado por cada jugador.
player_deposits = df_deposits.groupby('player_id')['deposit_amount'].sum().reset_index()

# Se ordena el DataFrame "player_deposits" de forma descendente por la columna "deposit_amount",
# para que los jugadores con mayor depósito aparezcan primero.
player_deposits = player_deposits.sort_values(by='deposit_amount', ascending=False)

# Se selecciona el primer registro del DataFrame ordenado, correspondiente al jugador con el mayor depósito.
top_player = player_deposits.iloc[0]

# Se imprime el ID del jugador con más dinero depositado y el valor total depositado, formateado con dos decimales.
print(f"Jugador con más dinero depositado: ID {top_player['player_id']} con un total de ${top_player['deposit_amount']:.2f}")
```

```python
# Se identifican los 10 primeros registros de la tabla, equivalente a los 10 jugadores con mayores depósitos
top_10_players = player_deposits.head(10)

# Se crea el gráfico de barras usando seaborn
plt.figure(figsize=(12, 6))
ax = sns.barplot(y=top_10_players['player_id'].astype(str), x=top_10_players['deposit_amount'], color='orange')

# Se personaliza el gráfico
plt.title('Top 10 Jugadores con Mayor Depósito')
plt.ylabel('Jugador')
plt.xlabel('Total Depositado (USD)')

# Se muestran los valores representativos encima de cada barra
for p in ax.patches:
    # La coordenada x es el valor del depósito, y la coordenada y es el ID del jugador
    ax.text(p.get_width() + 5, p.get_y() + p.get_height() / 2, f'{p.get_width():.0f}', 
            horizontalalignment='center', verticalalignment='center', fontsize=10, color='black')

# Se ajusta el entorno gráfico
plt.tight_layout()

# Se muestra el entorno gráfico
plt.show()
```

**Observaciones claves:**
- El mayor valor de depósitos realizados asciende una cifra de 2,156 USD.
- La diferencia entre el jugador con más dinero invertido y el segundo es de aproximadamente 309 USD.
- El top 10 de los jugadores con más depósitos, hechos han fondeado los servicios gambling con más de 1400 USD cada uno.

Para realizar una extracción más clara de insights, se calculan a continuación los valores de tendencia central de los depósitos.
